### Generator Function using yield

```python
def stream_clicks(big_data_source):
    for item in big_data_source:
        try:
            yield int(item) # This "pauses" and returns one int
        except (ValueError, TypeError):
            continue # Skip bad data

# Usage:
raw_clicks = [150, "200", None, 300, "invalid_data", 50, "100"]
total = sum(stream_clicks(raw_clicks))
print(total)
```

### The generator expression


```python
#Tip to remember - similar to list comprehension
#The parentheses () make it a generator instead of a list []
total = sum(int(x) for x in raw_clicks if str(x).isdigit())
print(total)
```
