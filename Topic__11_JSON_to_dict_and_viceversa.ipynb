### Converting json to dictionary:
{
  "name": "Aris",
  "role": "Collaborator",
  "status": "Online",
  "version": 3.0
}

```python
import json

# JSON string
json_data = '{"name": "Aris", "role": "Collaborator", "status": "Online", "version": 3.0}'

# Convert to dictionary
data_dict = json.loads(json_data)

print(f"Type: {type(data_dict)}")
print(f"Name: {data_dict['name']}")
```

### Converting json array to list of dictionaries

[
  {"id": 1, "task": "Coffee"},
  {"id": 2, "task": "Code"},
  {"id": 3, "task": "Sleep"}
]

```python
import json

# JSON array string
json_array = '[{"id": 1, "task": "Coffee"}, {"id": 2, "task": "Code"}, {"id": 3, "task": "Sleep"}]'

# Convert to a Python list
data_list = json.loads(json_array)

print(f"Type: {type(data_list)}")

# Accessing the second item's task
print(f"Second Task: {data_list[1]['task']}")
```

### Converting the dictionary to a string


```python
import json

# 1. A simple Python dictionary
coffee_order = {
    "type": "Espresso",
    "sugar": 0,
    "extra_hot": True
}

# 2. Convert (serialize) to a JSON string
# indent=2 makes it look like a "block" instead of one long line
json_output = json.dumps(coffee_order, indent=2)

print(json_output)
print(type(json_output))
```

```python
import json

# A Python dictionary with various data types
project_data = {
    "project_id": 99,
    "tags": ["python", "json", "tutorial"],
    "is_public": False,
    "rating": None
}

# Convert to a 'Pretty' string
# 'indent' adds spaces, 'sort_keys' alphabetizes the dictionary
json_string = json.dumps(project_data, indent=4, sort_keys=True)

print("Check out this formatted JSON string:")
print(json_string)
```
